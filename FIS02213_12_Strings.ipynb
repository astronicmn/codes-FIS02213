## Fatiamento de strings

```python
s = "abcdefghijklm"
#    0           1
#    012|3456|789012

print( s[0] )   # acesso ao elemento 

print( s[3:7] ) # opera√ß√£o de fatiamento
print( s[:5]  )
print( s[3:]  , end='\n\n')

print( s[2:10:2] , end='\n\n')

print( s[::-1] , end='\n\n')
```

## M√©todos de strings

### 1) M√©todos para opera√ß√µes com min√∫sculas e mai√∫sculas


```python
s = "proGRama√á√Éo EM linGUagEm pytHOn"

print( " s = " + s )

print( "s.lower()      : " + s.lower() )           # convers√£o para min√∫sculas
print( "s.upper()      : " + s.upper() )           # convers√£o para ma√≠√∫sculas
print( "s.capitalize() : " + s.capitalize() )      # primeira letra : mai√∫scula
print( "s.title()      : " + s.title() )           # t√≠tulo
print( "s.swapcase()   : " + s.swapcase() )        # inverte ma√≠usculas / min√∫sculas

print( " s = " + s )
```

### 2) M√©todos de particionamento

```python
# .split()

data = "25/07/2022"
print( data.split("/") )

print( "/home/usuario/dir1/file.dat".split("/") )

print( "Albert Einstein".split(" ") )
print( "Albert Einstein".split() ) # por default, espa√ßo √© usado como sperador
```

```python
# .partition() 
print( "abcdefg".partition("d") )
print( "abcdefg".partition("cde") )
print( "abcdefg".partition("a") )
```

```python
# .splitlines()

texto = '''Linha 1
Linha 2
Linha 3'''

print( texto.splitlines() )

texto2 = "Linha 1\nLinha 2\nLinha 3"
print(texto2.splitlines() )
```

### 3) M√©todos para justifica√ß√£o (posicionamento)

```python
s = "Instituto de F√≠sica"
print("*" + s  + "*")

print("*" + s.center(30)  + "*")   # posiciona no centro do campo
print("*" + s.ljust(30)  + "*")
print("*" + s.rjust(30)  + "*")

print(len(s))
```

### 4) M√©todos para remo√ß√£o de espa√ßos (em branco)

```python
s = "       Instituto de F√≠sica     "
print("*" + s  + "*")

print("*" + s.strip()  + "*")
print("*" + s.lstrip()  + "*")
print("*" + s.rstrip()  + "*")
```

### 7) M√©todos para pesquisa em strings

```python
# .count()

s = "abc abc abc xyz"

print( s.count("a") )
print( s.count("a", 2) )
print( s.count("a", 2, 6) )

print( s.count("abc") )
```

```python
# .find()
s = "abc abc abc xyz"
print( s.find("c") )
print( s.find("c", 2, 6) )

print( s.rfind("c") )

print( s.find("W") ) # retorna -1
```

```python
# .index()

s = "abc abc abc xyz"

print( s.index("c") )
print( s.index("c", 2, 6) )

print( s.rindex("c") )

print( s.index("W") ) # gera uma exce√ß√£o
```

### 8) M√©todos para testes

```python
# para testar se os caracteres de uma string s√£o d√≠gitos/n√∫mericos:

print( '123'.isdigit() )  # ASCII
print( '123.4'.isdigit() ) 
print()

print( '123'.isdecimal() )  # Unicode
print( '123.4'.isdecimal() ) 
print()

print( "\U00002154".isnumeric() )  # Unicode
```

```python
print( 'abc'.isalpha() ) 
print( 'abc123'.isalpha() ) 
print( 'abc 123'.isalpha() ) 
print()

print( 'abc'.isalnum() ) 
print( 'abc123'.isalnum() ) 
print( 'abc 123'.isalnum() ) 
print()

print( '    '.isspace() )
print( '    x'.isspace() )
print()

print( 'abc'.isprintable() )
print( 'abc \n def'.isprintable() )


```

```python
print( 'abcd'.islower() )
print( 'abcd EFG'.islower() )
print( 'EFG'.islower() )
print()

print( 'abcd'.isupper() )
print( 'abcd EFG'.isupper() )
print( 'EFG'.isupper() )
print()

print( 'Programa√ß√£o em Python'.istitle() )
print( 'Programa√ß√£o Em Python'.istitle() )
print( 'Programa√ß√£o em PYTHON'.istitle() )
```

```python
print( "Python".startswith("Py") )
print( "Python".startswith("py") )
print()

print( "Python".endswith("Py") )
print( "Python".endswith("on") )
print()
```

```python
print( 'x'.isidentifier() ) 
print( '_abc'.isidentifier() ) 
print( '.abc'.isidentifier() ) 
print( '12abc'.isidentifier() ) 
print( ''.isidentifier() ) 
print( 'print'.isidentifier() )
```

### 9) M√©todos para formata√ß√£o e interpola√ß√£o de strings

```python

# M√âTODO 1: usando o operador de interpola√ß√£o %      [Python]

# Suponha que se deseje repetir o padr√£o de mensagem abaixo para diferentes elementos
# qu√≠micos, mudando o s√≠mbolo, o nome e o n√∫mero at√¥mico:
print("O s√≠mbolo H corresponde ao elemento qu√≠mico Hidrog√™nio, cujo n√∫mero at√¥mico √© 1")
print()


# Defini-se a string padr√£o indicando com %<tipo> os locais onde os valores devem ser
# inseridos. O especificador <tipo> indica o tipo de valor (s=string, i=inteiro, f=float)
linha = "O s√≠mbolo %s corresponde ao elemento qu√≠mico %s, cujo n√∫mero at√¥mico √© %i"

print( linha%("H", "Hidrog√™nio", 1) )
print( linha%("He", "H√©lio", 2) )
print( linha%("C", "Carbono", 6) )

print('\n\n')
```

```python
# M√âTODO 2: usando o m√©todo .format()     [Python 2.6]

linha = "O s√≠mbolo {simb} corresponde ao elemento qu√≠mico {nome}, cujo n√∫mero at√¥mico √© {N}."


print( linha.format(simb="H", nome="Hidrog√™nio", N=1) )
print( linha.format(simb="He", nome="H√©lio", N=2) )
print( linha.format(nome="Carbono", N=6, simb="C") )       # <-- alterando a ordem dos argumentos
```

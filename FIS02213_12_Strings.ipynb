## Definindo strings

```python
# definindo uma string com vários caracteres:

s = 'Linguagem Python de Programação'

print(s, end='\n\n')
```

```python
# string com um único caractere:
s1 = "x"
print(s1, end='\n\n')
```

## Tamanho de strings

```python
# O número de caracteres de uma string é seu tamanho.
# A função len(string) retorna o tamanho de uma string.
```

```python

str_a = ''
str_b = 'x'
str_c = 'Linguagem Python'

print( len(str_a) )
print( len(str_b) )
print( len(str_c), end='\n\n' )
```

## Caractere de escape 

```python
# new line: \n
s = "O curso FIS02213 está sendo ministrado \nem modo de \"Ensino Remoto Emergencial\"."

print(s, end='\n\n')
```

```python
# tabulador: \t

print("Elemento \tSímbolo \tQuantidade")
print("Hidrogênio \tH \t\t12.2")
print("Hélio \t\tHe \t\t20.1", end='\n\n')
```

```python
# backspace: \b
print("01233367\b\b\b\b4567", end='\n\n')
```

```python
# emojis
print("\U0001F44D\n", "\U0001F61C \n", 6 * "\U0001F923", end='\n\n')
```

## Fatiamento de strings

```python
s = "abcdefghijklm"
#    0           1
#    012|3456|789012

print( s[0] )   # acesso ao elemento 

print( s[3:7] ) # operação de fatiamento
print( s[:5]  )
print( s[3:]  , end='\n\n')

print( s[2:10:2] , end='\n\n')

print( s[::-1] , end='\n\n')
```

## Métodos de strings

### 1) Métodos para operações com minúsculas e maiúsculas


```python
s = "proGRamaÇÃo EM linGUagEm pytHOn"

print( " s = " + s )

print( "s.lower()      : " + s.lower() )           # conversão para minúsculas
print( "s.upper()      : " + s.upper() )           # conversão para maíúsculas
print( "s.capitalize() : " + s.capitalize() )      # primeira letra : maiúscula
print( "s.title()      : " + s.title() )           # título
print( "s.swapcase()   : " + s.swapcase() )        # inverte maíusculas / minúsculas

print( " s = " + s )
```

### 2) Métodos de particionamento

```python
# .split()

data = "25/07/2022"
print( data.split("/") )

print( "/home/usuario/dir1/file.dat".split("/") )

print( "Albert Einstein".split(" ") )
print( "Albert Einstein".split() ) # por default, espaço é usado como sperador
```

```python
# .partition() 
print( "abcdefg".partition("d") )
print( "abcdefg".partition("cde") )
print( "abcdefg".partition("a") )
```

```python
# .splitlines()

texto = '''Linha 1
Linha 2
Linha 3'''

print( texto.splitlines() )

texto2 = "Linha 1\nLinha 2\nLinha 3"
print(texto2.splitlines() )
```

### 3) Métodos para justificação (posicionamento)

```python
s = "Instituto de Física"
print("*" + s  + "*")

print("*" + s.center(30)  + "*")   # posiciona no centro do campo
print("*" + s.ljust(30)  + "*")
print("*" + s.rjust(30)  + "*")

print(len(s))
```

### 4) Métodos para remoção de espaços (em branco)

```python
s = "       Instituto de Física     "
print("*" + s  + "*")

print("*" + s.strip()  + "*")
print("*" + s.lstrip()  + "*")
print("*" + s.rstrip()  + "*")
```

### 5) Métodos para substituição de substrings

```python
s = "Linguagem QiMthon"

print( s.replace("QiM", "Py") )

s = "   Ins ti tu to    "

print( s.replace(" ", "") )
```

### 7) Métodos para pesquisa em strings

```python
# .count()

s = "abc abc abc xyz"

print( s.count("a") )
print( s.count("a", 2) )
print( s.count("a", 2, 6) )

print( s.count("abc") )
```

```python
# .find()
s = "abc abc abc xyz"
print( s.find("c") )
print( s.find("c", 2, 6) )

print( s.rfind("c") )

print( s.find("W") ) # retorna -1
```

```python
# .index()

s = "abc abc abc xyz"

print( s.index("c") )
print( s.index("c", 2, 6) )

print( s.rindex("c") )

print( s.index("W") ) # gera uma exceção
```

### 8) Métodos para testes

```python
# para testar se os caracteres de uma string são dígitos/númericos:

print( '123'.isdigit() )  # ASCII
print( '123.4'.isdigit() ) 
print()

print( '123'.isdecimal() )  # Unicode
print( '123.4'.isdecimal() ) 
print()

print( "\U00002154".isnumeric() )  # Unicode
```

```python
print( 'abc'.isalpha() ) 
print( 'abc123'.isalpha() ) 
print( 'abc 123'.isalpha() ) 
print()

print( 'abc'.isalnum() ) 
print( 'abc123'.isalnum() ) 
print( 'abc 123'.isalnum() ) 
print()

print( '    '.isspace() )
print( '    x'.isspace() )
print()

print( 'abc'.isprintable() )
print( 'abc \n def'.isprintable() )


```

```python
print( 'abcd'.islower() )
print( 'abcd EFG'.islower() )
print( 'EFG'.islower() )
print()

print( 'abcd'.isupper() )
print( 'abcd EFG'.isupper() )
print( 'EFG'.isupper() )
print()

print( 'Programação em Python'.istitle() )
print( 'Programação Em Python'.istitle() )
print( 'Programação em PYTHON'.istitle() )
```

```python
print( "Python".startswith("Py") )
print( "Python".startswith("py") )
print()

print( "Python".endswith("Py") )
print( "Python".endswith("on") )
print()
```

```python
print( 'x'.isidentifier() ) 
print( '_abc'.isidentifier() ) 
print( '.abc'.isidentifier() ) 
print( '12abc'.isidentifier() ) 
print( ''.isidentifier() ) 
print( 'print'.isidentifier() )
```

### 9) Métodos para formatação e interpolação de strings

```python

# MÉTODO 1: usando o operador de interpolação %      [Python]

# Suponha que se deseje repetir o padrão de mensagem abaixo para diferentes elementos
# químicos, mudando o símbolo, o nome e o número atômico:
print("O símbolo H corresponde ao elemento químico Hidrogênio, cujo número atômico é 1")
print()


# Defini-se a string padrão indicando com %<tipo> os locais onde os valores devem ser
# inseridos. O especificador <tipo> indica o tipo de valor (s=string, i=inteiro, f=float)
linha = "O símbolo %s corresponde ao elemento químico %s, cujo número atômico é %i"

print( linha%("H", "Hidrogênio", 1) )
print( linha%("He", "Hélio", 2) )
print( linha%("C", "Carbono", 6) )

print('\n\n')
```

```python
# MÉTODO 2: usando o método .format()     [Python 2.6]

linha = "O símbolo {simb} corresponde ao elemento químico {nome}, cujo número atômico é {N}."


print( linha.format(simb="H", nome="Hidrogênio", N=1) )
print( linha.format(simb="He", nome="Hélio", N=2) )
print( linha.format(nome="Carbono", N=6, simb="C") )       # <-- alterando a ordem dos argumentos
```

```python
#!/usr/bin/env python3 

'''
	Informa se um ano é bissexto ou normal.
   
	DESCRIÇÃO:
		Um ano é bissexto se ele é divisível por 4;
		não divisível por 100, exceto se for divisível por 400.
		 
		Válido para anos a partir de 1582

	AUTOR:
		N. M. Narvaz
		00318736@ufrgs.br
	
	REFERÊNCIAS: 
		<[#] referência..>
	
	BUGS/LIMITAÇÕES:

	ToDos:
		<alterações/correções a fazer>
	
	OBSERVAÇÕES:
		Para tornar executável:
		% chmod ugo+x FIS02213_09_bissexto.py
		
		Para criar um link em ~/bin:
		% ln ~/prog/python/FIS02213_09_bissexto.py ~/bin/bissexto
	 
	VERSÃO:
	0.01 - <11.07.2022> 
'''

import sys

#Entrar o ano:

ano = int(input("Digite um ano (ano > 1582) : ") )

#Validar ano:
if ano <= 1582:
	print(f"Ano inválido.")
	sys.exit() #interrompe o programa.

#Modo 1: ifs alinhados
if ano%4==0:
	if ano%100==0:
		if ano%400==0:
			print(f"{ano} é bissexto.")
		else:
			print(f"{ano} é normal.")
	else:
		print(f"{ano} é bissexto.")
else:
	print(f"{ano} é normal.")

#Modo 2: if-elif-else
if ano%4 != 0:
	print(f"{ano} é normal.")
elif ano%400==0:
	print(f"{ano} é bissexto.")
elif ano%100==0:
	print(f"{ano} é normal.")
else:
	print(f"{ano} é bissexto.")
#Modo 3:
if ano%4 == 0 and (ano%100 != 0 or ano%400==0):
	print(f"{ano} é bissexto.")
else:
	print(f"{ano} é normal.")

#Modo 4:
tipo = 'bissexto' if (ano%4 == 0 and (ano%100 != 0 or ano%400==0)) else 'normal'
print(f"{ano} é {tipo}")

#Modo 5:
print(f"{ano} é {'bissexto' if (ano%4 == 0 and (ano%100 != 0 or ano%400==0)) else 'normal'}.")
```

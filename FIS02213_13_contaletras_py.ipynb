```python
#!/usr/bin/env python3 

'''
Contabiliza o número de letras do alfabeto.
   
DESCRIÇÃO:
    Contabiliza o número de ocorrências das letras do alfabeto
     em um dado texto.

AUTOR:
    N. M. Narvaz
    00318736@ufrgs.br

REFERÊNCIAS: 
    <[#] referência..>

BUGS/LIMITAÇÕES:

ToDos:
    <alterações/correções a fazer>

OBSERVAÇÕES:
    Para tornar executável:
    % chmod ugo+x FIS02213_13_contaletras.py

    Para criar um link em ~/bin:
    % ln ~/prog/python/FIS02213_13_contaletras.py ~/bin/contaletras

VERSÃO:
    0.01 - 25.07.2022 
'''

texto = input("Digitar texto: ")

alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

contador = 0

for letra in alfabeto:
    n = texto.upper().count(letra) #texto convertido para letras maiúsculas
    print(f"\t{letra}: \t{n}\t|", end = '')
    
    contador+=1
    
    if contador == 4:
        print()
        contador=0
        
print()
```

```python
# Importando módulos:
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Solicitando valores t1 e t2
print('Insira os valores do intervalo de tempo Δt t1 e t2, lembrando que t2 > t1!, \n')
while True: # enquanto a sentença for verdadeira.
    try:    # repete ao usuário se a condição não for cumprida. 
        t1 = float(input('Insira o tempo inicial t1: ')) # valores em float.
        t2 = float(input('Insira o tempo final t2: '))   # valores em float.
        if t2 > t1: # confirma a condição.
            break   # quando confirmada, continua o código.
        else:       # se a condição não for cumprida, retorna ao usuário um erro.
            print('O valor inserido não pode ser t2 < t1! Digite novamente o valor sendo que t2 > t1.')
    except:         # se o usuário digitar um valor não númerico, retorna um erro.
         print('Você inseriu um valor inválido, digite novamente.')
            
print('Feito!') # valores inseridos corretamente.
```

```python
# Solicitando número de medidas n
print('Insira o número n de medidas realizadas durante este intervalo de tempo, sendo n > 1 e inteiro!, \n')

while True: # enquanto a sentença for verdadeira.
    try:    # repete ao usuário se a condição não for cumprida. 
        n = int(input('Insira o número de medidas n: ')) # valor inteiro.
        if n <= 1: # confirma a condição.
            print('O valor inserido não pode ser n < 1! Digite novamente o número de medidas sendo n > 1.')
            continue # se valor inserido não cumprir a condição retorna erro.
        else:        # se valor cumprir a condição,
            break    #  continua código e recebe novo valor
    except:          # se o usuário digitar um valor não númerico, retorna um erro.
        print('Você inseriu um valor inválido, digite novamente.')
        
print('Feito!') # valores inseridos corretamente.
```

```python
# Solicitando posição inicial x0
print('Insira a posição inicial x0, a velocidade inicial v0 e a aceleração da partícula a!, \n')
while True:
    try:
        x0 = float(input('Insira a posição inicial x0: '))
        break
    except:
        print('Você inseriu um valor inválido, digite novamente.')
print('Feito!') # valores inseridos corretamente.

# Solicitando velocidade inicial v0
while True:
    try:
        v0 = float(input('Insira a velocidade inicial v0: '))
        break
    except:
        print('Você inseriu um valor inválido, digite novamente.\n')
print('Feito!') 

# Solicitando aceleração da partícula a
while True:
    try:
        a = float(input('Insira a aceleração da partícula a: '))
        break
    except:
        print('Você inseriu um valor inválido, digite novamente.')
print('Feito!')
```

```python
# Definindo a equação cinemática para o andamento do código:
def mruv(t):
    xf = x0 + v0*t + 1/2*a*t**2
    return xf 

t = np.linspace(t1, t2, n) # gera as medidas realizadas durante o intervalo de tempo.
x = mruv(t)                # gera as posições em relação as medidas. 

# Retornar os dados em lista para melhor visualização:
lista_tempo = t.tolist() # tolist retorna array como lista
lista_posicao = x.tolist()
```

```python
datafile = (input('Insira o nome do arquivo: ')+'.txt') # nome do arquivo datafile e como .txt (texto)
with open(datafile, 'w') as file: 
    file.write("Tempo \t Posição\n") 
    for t, x in zip(lista_tempo, lista_posicao):
        file.write(f"{t} \t{x}\n") 
print('Sucesso, arquivo salvo (no mesmo diretório do código) como', datafile)
```

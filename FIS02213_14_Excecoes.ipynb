## Captura de Exceções

```python
try:
  x = float(input("Digite um número real: "))

except ValueError:
  print("Você digitou um número errado!")

print("Terminei!")
```

```python
# Exemplo 2

try:
  x = float(input("Digite um número real:"))

except Exception:
  print("Você digitou um número errado!")

print("Terminei!")
```

```python
# Exemplo 3
import math

try:
  x = float(input("Digite x: "))
  y = 1 / math.sqrt(x)

except ValueError:
  print("Você digitou um valor errado!")

except ZeroDivisionError:
  print("Você digitou zero!")

print(x, y)
print("Terminei!")
```

```python
# Exemplo 4
import math

try:
  x = float(input("Digite x: "))
  y = 1 / math.sqrt(x)

except ValueError:
  print("Você digitou um valor errado!")

except ZeroDivisionError:
  print("Você digitou zero!")

else:
  print(x, y)

print("Terminei!")
```

```python
# Exemplo 6
import math

try:
  x = float(input("Digite x: "))
  y = 1 / math.sqrt(x)

except ZeroDivisionError:
  print("Você digitou zero!")

except Exception:    # ele pula esta cláusula porque a de cima já foi entendida
  print("Você digitou um valor errado!")

else:
  print(x, y)

print("Terminei!")
```

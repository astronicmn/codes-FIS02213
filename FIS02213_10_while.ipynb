## Exemplo 1:
Imprimir todos os 10 primeiros números naturais.

```python
n = 0

while n<10:
    n+=1
    print(n)
```

## Exemplo 3:
O usuário deve digitar números inteiros até que o usuário digite 0.
Informar qual foi o maior número digitado pelo usuário.

```python
while True:
    n = int( input("Digite um número (0: parar): "))
    if n == 0:
        break
```

```python
maior = 0
while True:
    n = int( input("Digite um número (0: parar): "))
    if n > maior:
        maior=n
    if n == 0:
        break
print(f"O maior número é {maior}")
```

## Exemplo 4:
O código deve fazer a validação de uma senha.
O usuário digita senha1.
o usuário digita novamente a senha --> senha2.
A senha digitada é válida se senha1==senha2.

```python
while True:
    senha1 = input("Digite uma senha: ")
    senha2 = input("Confirme a senha: ")
    
    if senha1 == senha2:
        print("Senha válida.")
        break
    else:
        print("Senha inválida. Tente novamente", end='\n\n')
```

## Exemplo 5:
Criando um menu de opções:

```python
import math 

x = float( input("Digite x: "))

while True:
    print()
    print("[a] sqrt(x)")
    print("[b] log10(x)")
    print("[c] exp(x)")
    print("[q] quit")
    opcao = input("Opção: ")
    
    if opcao in ['a', 'b', 'c', 'q']:
        break
        
if opcao == 'a':
    print(math.sqrt(x))
elif opcao == 'b':
    print(math.log10(x))
else:
    print(math.exp(x))
```

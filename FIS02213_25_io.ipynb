```python
dirpath = '/home/nicmn/prog/python/FIS02213/data/'
filename = dirpath + 'teste.dat'

#%% Abrindo e fechando um arquivo

fid = open(filename, 'w') # w : write 
print(fid, type)
print(fid.closed)
fid.close()

#%% Salvando dados em um arquivo

fid = open(filename, 'w')
fid.write('linha 1\n')
fid.write('linha 2\n')
fid.write('linha 3\n')
fid.close()

#%% Salvando dados em um arquivo part.2

lista = ['linha A\n', 'linha B\n', 'linha C\n']

fid = open(dirpath+'teste2.dat', 'w')
fid.writelines(lista)
fid.close()

#%% Salvando dados em um arquivo part.3

lista = ['linha A', 'linha B', 'linha C']
lista = [e +'\n' for e in lista] #cada elemento da lista sucedido    
                                # por nova linha 

fid = open(dirpath+'teste2.dat', 'w')
fid.writelines(lista)
fid.close()

#%% Salvando n√∫meros/ escrevendo dados em um arquivo

X = [0, 1, 2, 3, 4]
Y = [e**2 for e in X]

fid = open(dirpath+'teste3.dat', 'w')
for x, y in zip(X, Y): #pegando um valor da linha X e um valor da Y
    fid.write(f"{x} {y}\n")
fid.close()

#%% Lendo um arquivo

fid = open(dirpath+'teste2.dat', 'r') #r: read

#iterando o objeto
for linha in fid:
    print(linha)

fid.close()

#%% Lendo um arquivo part.2

fid = open(dirpath+'teste2.dat', 'r') #r: read

lista = fid.readlines()

fid.close()
print(lista)

#%% Lendo um arquivo com with

with open(dirpath+'teste2.dat', 'r') as fid:
    for linha in fid:
        print(linha)
#with automaticamente lida com o fechamento do arquivo

#%% Escrevendo dados em um arquivo com with

with open(dirpath+'teste4.dat', 'w') as fid:
    fid.write('linha 10\n')
    fid.write('linha 11\n')
    fid.write('linha 12\n')
```

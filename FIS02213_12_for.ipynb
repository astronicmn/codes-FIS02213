```python
#incremento negativo
for i in range(20, 0, -2):
    print(i)
```

```python
for i in range(0, 21, 3):
    if i==9:
        continue #passa diretamente p/ próxima interação
    print(i)
```

```python
# Usar a função range para gerar valores pares de 0 a 100 (incluindo o 100)
# Mostrar apenas os valores entre 30 e 70

#Solução 1:

for i in range(0, 101, 2):
    if i < 30:
        continue
    if i > 70:
        break
    print(i)
```

```python
#Cláusula else

for i in range(0, 10):
    print(i)
else:
    print("Loop finalizado")
```

```python
for i in range(0, 10):
    if i > 7:
        break
    print(i)
else:
    print("Loop finalizado")
```

```python
#Loops alinhados 

for linha in range(1, 4):
    for coluna in range(1, 4):
        for x in range(1, 4):
            print(f"a_ij({linha},{coluna})")
```

```python
for linha in range(1, 4):
    for coluna in range(1, 4):
        print(f"a_ij({linha},{coluna})")
```

```python
for linha in range(1, 4):
    for coluna in range(1, 4):
        for k in range(1, 4):
            print( f"a({linha},{coluna},{k})" )
```

```python
#!/usr/bin/env python3

# Demonstra o uso das funções de entrada e saída via terminal 
# to Python: input() e print()

# Importando módulos:
import cmath
from cmath import sqrt

# Entra valores de a, b e c:
a = float( input("Digite o valor de a: ") )
b = float( input("Digite o valor de b: ") )
c = float( input("Digite o valor de c: ") )

# Entra as raízes usando Bhaskara:
delta = b**2 - 4 * a * c

x1 = -b/(2*a) - cmath.sqrt(delta)/(2*a)
x2 = -b/(2*a) + cmath.sqrt(delta)/(2*a)

# Saída de dados:
print(f" A raíz x1 é igual a {x1}")
print(f" A raíz x2 é igual a {x2}")
```
